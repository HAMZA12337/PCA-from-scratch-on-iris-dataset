```python
## pandas,numpy,matplotlib,seabron,sklearn

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt;
import seaborn as sns
```

```python
# chargement des donn√©es
iris = pd.read_csv('iris.csv')
```

```python
# Explore the dataset
print(iris.head())
print("---------------------------------------------------")
print(iris.shape)
print("-----------------------------------")
# In this case, iris.describe() generates a table that includes the count, mean, standard deviation, minimum, 
# and the mean row shows the average value of 
# each column. The std(ecart type) row shows the standard deviation of each column, 
print(iris.describe())
```

```python
# Create scatterplot matrix to visualize the relationships between variables
sns.pairplot(data=iris, hue='class')
```

```python
# Create boxplot to visualize the distribution of each variable
iris.boxplot()


```

```python
# Create heatmap to visualize the correlations between variables
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')
```

```python
# Split the data into training and testing sets
# The train_test_split function of the sklearn.model_selection package 
# in Python splits arrays or matrices into random subsets for train and test data, respectively.
from sklearn.model_selection import train_test_split
X = iris.drop('class', axis=1)
y = iris['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```
